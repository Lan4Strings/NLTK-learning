## 4.1 Lexical Relations

```python
wn.synsets('motorcar') #find synsets
```

```python
wn.synset('car.n.01').lemma_names # list words 
wn.synset('car.n.01').definition #definition
wn.synset('car.n.01').examples #examples
wn.lemma('car.n.01.automobile').name
```

### 4.1.1 nouns belonging relations

```python
#meronyms: from top to bottom
#holonyms: from bottom to top
wn.synset('tree.n.01').part_meronyms() # tree has parts
```

```python
wn.synset('burl.n.02').part_holonyms() # burl is part of trees
```

```python
wn.synset('tree.n.01').substance_meronyms() # tree is made of heartwood and sapwood
```

```python
wn.synset('heartwood.n.01').substance_holonyms() # heartwood can made trees
```

```python
wn.synset('tree.n.01').member_holonyms() # tree is a member of forest
```

### 4.1.2 verbs belonging relations

```python
wn.synset('eat.v.01').entailments() # eat includes chew and swallow
```

### 4.1.3 Antonyms relations

```python
wn.lemma('supply.n.02.supply').antonyms()
```

### 4.1.4 find the relations between words

```python
orca = wn.synset('orca.n.01')
minke = wn.synset('minke_whale.n.01')
```

```python
orca.lowest_common_hypernyms(minke)
#they all belongs to whales
```

```python
orca.path_similarity(minke)
# the similarity between orca and minke 14%
```

# 8. Part-Of-Speech 

```python
#pos tag
nltk.pos_tag(text) # or nltk.pos_tag(tokens)
```

```python
#corpus tagger
nltk.corpus.nps_chat.tagged_words() #informal language
nltk.corpus.brown.tagged_words()
nltk.corpus.treebank.tagged_words()
nltk.corpus.sinica_treebank.tagged_words() #chinese
```

```python
#using regular expression for tagging
patterns = [(r'.*ing$', 'VBG'),
            (r'.*ed$', 'VBD'),
            (r'.*es$', 'VBZ'),
            (r'.*ould$', 'MD'),
            (r'.*\'s$', 'NN$'),
            (r'.*s$', 'NNS'),
            (r'^-?[0-9]+(.[0-9]+)?$', 'CD'), # cardinal numbers
            (r'.*', 'NN') # nouns (default)
           ]
regexp_tagger=nltk.RegexpTagger(patterns)
regexp_tagger.tag(text)
```

## 8.1 n-gram tagging

```python
#unigram taggers
from nltk.corpus import brown
brown_tagged_sents = brown.tagged_sents()
brown_sents = brown.sents(categories='news')

bigram_tagger = nltk.UnigramTagger(brown_tagged_sents)
bigram_tagger.tag(brown_sents[10])  #it should be run on the sentence tokens
```

```python
#bigram taggers
bigram_tagger = nltk.BigramTagger(brown_tagged_sents)
bigram_tagger.tag(brown_sents[10]) #it should be run on the sentence tokens
```
